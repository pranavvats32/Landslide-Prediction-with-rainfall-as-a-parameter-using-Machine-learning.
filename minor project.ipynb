```python
df['landslide2'] = labelencoder.fit_transform(df['landslide1'])
```

```python
df.groupby('landslide1').count()
```

```python
df.groupby('landslide2').count()
```

```python
X = df.copy()
Y = X['landslide2']
X.drop(['landslide2'], axis = 1, inplace=True)
X.drop(['the_geom','date no time','landslide_','landslide1'], axis = 1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state= 5)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_classifier = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
```

```python
df['State'] = labelencoder.fit_transform(df['State'])
df['trigger'] = labelencoder.fit_transform(df['trigger'])
```

```python
rf_classifier.fit(X_train, Y_train)
```

```python
Y_pred_rf = rf_classifier.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, Y_pred_rf)
```

```python
n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 15)]
max_features = ['auto', 'sqrt']
max_depth = [2,4]
min_samples_split = [2, 5]
min_samples_leaf = [1, 2]
bootstrap = [True, False]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
rf_Grid = GridSearchCV(estimator = rf_classifier, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)
```

```python
rf_Grid.fit(X_train, Y_train)
```

```python
rf_Grid.best_params_
```

```python
print (f'Train Accuracy - : {rf_Grid.score(X_train,Y_train):.3f}')
print (f'Test Accuracy - : {rf_Grid.score(X_test,Y_test):.3f}')
```
